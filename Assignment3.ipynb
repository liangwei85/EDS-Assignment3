```python
url1 = 'http://www.mudah.my/Malaysia/basikal-for-sale?o='
url2 = '&q=basikal&so=1&th=1'

# http://www.mudah.my/Malaysia/basikal-for-sale?o=3&q=basikal&so=1&th=1

#http://www.mudah.my/Malaysia/basikal-for-sale?lst=0&fs=1&w=3&cg=0&q=basikal&so=1&st=s
#http://www.mudah.my/Malaysia/Electronics-3000/iphone7-for-sale?lst=0&fs=1&w=3&cg=3000&q=iphone7&so=1&st=s

product_name=[]
prices=[]
time = []
state = []
Sale_type= []
for page in range(0,68):
    url_page = url1+str(page)+url2
    scrape = requests.get(url_page)
    soup = BeautifulSoup(scrape.content, 'lxml')
    link = soup.find_all('div',{'class':'top_params_col1'})
    link1 = soup.find_all('div',{'class':'ads_price'})
    link2 = soup.find_all('div',{'class':'top_params_col2'})
    link3 = soup.find_all('div',{'class':'area'})
    link4 = soup.find_all('div',{'class':'middle_params_col2'})
    
    length = len(link) 
    for i in range(0,length):
        name = link[i].h2.text
        price = link1[i].text.strip().replace('RM ',"").replace("per month","").replace(" ","")
        timestamp = link2[i].div.div.text
        location = link3[i].text.strip()
        Sale_type1 = link4[i].div.text
        
        product_name.append(name)
        prices.append(price)
        time.append(timestamp)
        state.append(location)
        Sale_type.append(Sale_type1)

#         price = link1[i].b.text
#         product_name.append(name)
#         prices.append(price)
```

```python
df = pd.DataFrame({'name':product_name,'price': prices, 'Time':time ,'State': state, 'Sale_Type': Sale_type})

df['price'] = pd.to_numeric(df['price'])

df['Time'] = df['Time'].str.replace("Today,", '25 Jul,')
df['Time'] = df['Time'].str.replace("Yesterday,", '24 Jul,')
df['Time'] = df['Time'].str.replace("," ,  " 2007,")
df['Time'] = pd.to_datetime(df['Time'])
df['Date'] = pd.to_datetime(df['Time'].dt.date)

df['Month'] = df['Time'].apply(lambda x: x.month)
df['Day'] = df['Time'].apply(lambda x: x.day)
df['Hour'] = df['Time'].apply(lambda x: x.hour)


# df['Day'] = df['Time'].str[0:3]
# df['Hour'] = df['Time'].str[-6:-4]
# df['Month'] = df['Time'].str[3:7]
# df['Day'] = pd.to_numeric(df['Day'])
# df['Hour'] = pd.to_numeric(df['Hour'])
```

### Surging Kuala Lumpur "Basikal" posts
We found that they are a lot of posts belonging to companies concentrating in Kuala Lumpur only.
Other states are also depicted the same but with less significant posts.

```python
def greater_than_50_pct(val):
    
#     Takes a scalar and returns a string with
#     the css property `'color: red'` for negative
#     strings, black otherwise.
    
    color = 'blue' if val >= 50 else 'gray'
    return 'color: %s' % color


barcount = df.groupby(['State',Sale_type]).price.count().unstack()
barcount.Private = barcount.Private.fillna(0)
barcount['Total'] = barcount.Company + barcount.Private
barcount['Pct_Company'] = barcount.Company / barcount.Total * 100
barcount['Pct_Private'] = barcount.Private / barcount.Total * 100
barcount['Pct_Company'] = barcount['Pct_Company'].round(1)
barcount['Pct_Private'] = barcount['Pct_Private'].round(1)
barcount.sort_values('Total', ascending=False, inplace=True)
barcount.style.applymap(greater_than_50_pct, subset=['Pct_Company','Pct_Private'])
```

```python
def greater_than_50_pct(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'blue' if val >= 80 else 'black'
    return 'color: %s' % color


barsum = df.groupby(['State',Sale_type]).price.sum().unstack()
barsum.Private = barsum.Private.fillna(0)
#barsum.sort_values('Time', ascending=False, inplace=True)
barsum['Total'] = barsum.Company + barsum.Private
barsum['Pct_Company'] = barsum.Company / barsum.Total * 100
barsum['Pct_Private'] = barsum.Private / barsum.Total * 100

barsum.sort_values('Total', ascending=False, inplace=True)

barsum['Pct_Company'] = barsum['Pct_Company'].round()
barsum['Pct_Private'] = barsum['Pct_Private'].round()
barsum.Company = barsum.Company.map(lambda x: '{:,}'.format(x))
barsum.Private = barsum.Private.map(lambda x: '{:,}'.format(x))
barsum.Total = barsum.Total.map(lambda x: '{:,}'.format(x))

barsum.style.applymap(greater_than_50_pct, subset=['Pct_Company','Pct_Private'])


```

## Company Kuala Lumpur Average Basikal Price = RM3,457
KL Companies drove the average price up to RM 3,457.

```python
def greater_than_50_pct(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'blue' if val >= 100 else 'black'
    return 'color: %s' % color

baraverage = df.groupby(['State',Sale_type]).price.mean().unstack()
baraverage.Private = baraverage.Private.fillna(0)
baraverage.sort_values('Company', ascending=False, inplace=True)
baraverage.Private = baraverage.Private.round()
baraverage.Company = baraverage.Company.round()

baraverage["Company vs Private Growth %"] = (baraverage.Company / baraverage.Private - 1) * 100
baraverage["Company vs Private Growth %"] = baraverage["Company vs Private Growth %"].round()
baraverage.sort_values("Company vs Private Growth %", ascending=False, inplace=True)

baraverage.style.applymap(greater_than_50_pct, subset=["Company vs Private Growth %"])
```
